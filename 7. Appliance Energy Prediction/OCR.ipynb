```python
from paddleocr import PaddleOCR
from PIL import Image
import cv2

# Initialize OCR engine (downloads models automatically on first run)
# Disable advanced features for basic handwriting extraction; set lang='en' for English
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Simple setup; models auto-download

# Load image (replace 'writing_paper.jpg' with your file path or URL)
img_path = 'Letter.png'
image = Image.open(img_path).convert('RGB')
image.save('temp.jpg')  # PaddleOCR expects this format

# Perform OCR
result = ocr.ocr('temp.jpg')

# Extract and print text
extracted_text = []
if result and result[0]:
    for line in result[0]:
        bbox = line[0]
        if isinstance(line[1], (tuple, list)) and len(line[1]) >= 2:
            text = line[1][0]
            conf = line[1][1]
        else:
            text = str(line[1])
            conf = 0.0
        extracted_text.append(text)
        print(f"Text: '{text}' (conf: {conf:.2f})")
        print(f"Box: {bbox}\n")
else:
    print("No text detected.")


# Optional: Visualize results
# from paddleocr import draw_ocr
# im_show = draw_ocr(image, result[0], font_path='path_to_font.ttf')  # Save as image
# Image.fromarray(im_show).save('result_with_boxes.jpg')

full_content = '\n'.join(extracted_text)
print("\nFull extracted content:\n", full_content)
```

```python
from paddleocr import PaddleOCR  # Only this import needed

# Initialize (downloads models on first run)
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Or use_doc_orientation_classify=False for simpler setup

# OCR on image
img_path = 'Letter.jpg'
result = ocr.ocr(img_path)  # Returns list of results

# Print text results
for idx in range(len(result)):
    res = result[idx]
    for line in res:
        print(line)  # Each line: [[bbox], (text, score)]

# Visualize and save automatically (no draw_ocr needed)
for idx in range(len(result)):
    result[idx].print()  # Prints to console
    result[idx].save_to_img('output')  # Saves visualized image to output/ folder
    result[idx].save_to_json('output')  # Saves JSON structured data
```
